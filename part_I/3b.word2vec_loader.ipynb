## Visualizing WordVectors

```python
tsne=TSNE(n_components=2,perplexity = 3, method ='barnes_hut',verbose=1)
```

```python
vocab=['brasil', 'gremio','inter','palmeiras', 'corinthians',
       'nerd','computacao', 'software','tecnologia', 'game',
       'coronavac', 'pfizer', 'vacina', 'astrazeneca']
```

```python
tokens = []
for word in vocab:
    tokens.append(model[word])
X_tsne = tsne.fit_transform(tokens)
```

```python
plt.figure(figsize=(20, 10), dpi=150, facecolor='w', edgecolor='k')
plt.scatter(X_tsne[:,0], X_tsne[:,1])
for label, x, y in zip(vocab,X_tsne[:,0], X_tsne[:,1]):
    font = {'size' : 18, 'weight' : 'normal'}
    plt.rc('font', **font)
    plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
plt.grid(True)
plt.show()
```
